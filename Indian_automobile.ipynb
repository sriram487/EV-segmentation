```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv("Indian automoble buying behavour study 1.0.csv")
```

```python
data.head()
```

```python
data.Salary.value_counts(normalize=True)
```

```python
data.Salary.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
data.Wife_Salary.value_counts(normalize=True)
```

```python
data.Wife_Salary.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
data.Total_Salary.value_counts(normalize=True)
```

```python
data.Total_Salary.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
sns.pairplot(data = data, vars=['Total_Salary','Price','Age'])
plt.show()
```

```python
data[['Age','Total_Salary','Price']].corr()


sns.heatmap(data[['Age','Total_Salary','Price']].corr(), annot=True, cmap = 'Reds')
plt.show()
```
